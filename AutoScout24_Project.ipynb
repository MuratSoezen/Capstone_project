# columns I can drop from  DataFrame


To determine which columns you can drop from a DataFrame, you need to consider the specific characteristics of your dataset and the analysis or modeling task at hand. However, here are some common scenarios where dropping columns may be appropriate:
Irrelevant or unnecessary columns: If a column does not provide any relevant information or is not needed for your analysis, you can drop it. Examples include columns with constant values, unique identifiers, or descriptive information that is not relevant to your analysis.

Redundant or highly correlated columns: If two or more columns provide essentially the same information or are highly correlated, keeping only one of them can reduce redundancy and improve efficiency. This is particularly important in cases of multicollinearity in regression analysis.

Missing data: If a column has a significant amount of missing values and those missing values cannot be reasonably imputed or filled, you may consider dropping the column. However, it's important to carefully evaluate the impact of dropping a column with missing data on the integrity and representativeness of your dataset.

High cardinality categorical variables: If a categorical column has a large number of unique categories, and those categories do not contribute significantly to your analysis or modeling task, you may choose to drop that column. High cardinality can introduce complexity and potential overfitting in certain algorithms.

Columns with excessive outliers: If a column contains a significant number of outliers that are inconsistent with the distribution of the majority of the data, removing the column may be considered. Outliers can adversely affect the performance of some models and skew analysis results.

Remember that these guidelines are general and may vary depending on your specific dataset and analysis objectives. It's crucial to carefully evaluate the characteristics of your data, consider the impact of dropping columns on the overall analysis, and make informed decisions based on your specific requirements.

```python
Benzine= ['Super 95','Regular/Benzine 91','Super E10 95','Super E10 95','Regular/Benzine 91 (Particle filter)',
         'Super 95 (Particle filter)','Super E10 95 (Particle filter)','Regular/Benzine E10 91','Super Plus 98',
         'Super Plus E10 98','Super Plus E10 98 (Particle filter)','Super Plus 98 (Particle filter)',
          'Regular/Benzine E10 91 (Particle filter)','Super 95 (Particle filter) / Super E10 95 / Ethanol']

LPG=['Liquid petroleum gas (LPG)','CNG','LPG','Domestic gas H','Domestic gas L','CNG (Particle filter)',
     'LPG (Particle filter)','Domestic gas L (Particle filter)','Liquid petroleum gas (LPG) (Particle filter)',
    'Biogas']

Diesel=['Gasoline','Diesel','Diesel (Particle filter)','Gasoline (Particle filter)']

Benzine_LPG=['Liquid petroleum gas (LPG) (Particle filter) / Super 95 / Super E10 95',
             'Liquid petroleum gas (LPG) / Super 95 / Super E10 95',
            'Liquid petroleum gas (LPG) / Super 95 / Super Plus 98 / Super Plus E10 98 / Super E10 95',
             'Liquid petroleum gas (LPG) / Super 95',
             'Liquid petroleum gas (LPG) / Super 95 / Super E10 95 / Super Plus 98',
             'Domestic gas H / Super E10 95 / Super Plus E10 98 / Super 95 / Super Plus 98 / Domestic gas L',
             'Domestic gas L / Super 95 / Domestic gas H','Super 95 / Super Plus 98 / Liquid petroleum gas (LPG)',
            'Liquid petroleum gas (LPG) / Super 95 / Super Plus 98','Liquid petroleum gas (LPG) / Super E10 95',
             'Liquid petroleum gas (LPG) / Super E10 95 / Super Plus E10 98 / Super Plus 98 / Super 95',
             'Liquid petroleum gas (LPG) / Super E10 95 / Regular/Benzine 91 / Super 95 / Super Plus 98 / Biogas',
            'Super 95 / Liquid petroleum gas (LPG)']

Electric=['Electric','Electric (Particle filter)']

def fueltype(x):
    if x in Benzine:
        return 'Benzine'
    elif x in LPG: 
        return 'LPG/CNG und other gas'
    elif x in Diesel:
        return 'Diesel'
    elif x in Electric:
        return 'Electric'
    elif x in Benzine_LPG:
        return 'Benzine_LPG'
    else:
        return x
```
